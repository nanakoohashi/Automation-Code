```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import zipfile
```

```python
with zipfile.ZipFile('ACSST5Y2018.S1810_2020-02-13T121501.zip', 'r') as zip_ref:
    zip_ref.extractall()
```

```python
df = pd.read_csv('ACSST5Y2018.S1810_data_with_overlays_2020-02-13T121455.csv')
df.head()
```

```python
#remove header row
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
```

```python
df.head()
```

```python
df = df[df.columns.drop(list(df.filter(regex='Margin')))]
df.head()
```

```python
df.columns = df.columns.str.replace('Estimate!!Percent with a disability!!Subject!!DISABILITY TYPE BY DETAILED AGE!!', '')
df.head()
```

```python
df = df[df.columns.drop(list(df.filter(regex='under 5 years')))]
df = df[df.columns.drop(list(df.filter(regex='5 to 17 years')))]
df = df[df.columns.drop(list(df.filter(regex='18 to 64 years')))]
df = df[df.columns.drop(list(df.filter(regex='65 years and over')))]
df = df[df.columns.drop(list(df.filter(regex='65 to 74 years')))]
df = df[df.columns.drop(list(df.filter(regex='75 years and over')))]
df = df[df.columns.drop(list(df.filter(regex='35 to 64 years')))]
df = df[df.columns.drop(list(df.filter(regex='AGE!!')))]
df = df[df.columns.drop(list(df.filter(regex='RACE')))]
df = df[df.columns.drop(list(df.filter(regex='SEX')))]
df = df[df.columns.drop(list(df.filter(regex='DISABILITY TYPE BY DETAILED AGE!!')))]
df.head()
```

```python
df = df.loc[:, df.columns.str.contains('under 18|Area')]
df.head()
```

```python
df = df.rename(columns={col: col.split('!!')[0] for col in df.columns})
df.columns
```

```python
df.head()
```

```python
df = df.rename(columns={"Geographic Area Name": "Area"})
df.head()
```

```python
df = df.iloc[::-1]
df.head()
```

```python
df['With a hearing difficulty'] = df['With a hearing difficulty'].astype(str) + '%'
df['With a vision difficulty'] = df['With a vision difficulty'].astype(str) + '%'
df['With a cognitive difficulty'] = df['With a cognitive difficulty'].astype(str) + '%'
df['With an ambulatory difficulty'] = df['With an ambulatory difficulty'].astype(str) + '%'
df['With a self-care difficulty'] = df['With a self-care difficulty'].astype(str) + '%'
```

```python
df.head()
```

```python
df.to_csv('Types of Disabilities.csv', index=False)
```
